Overall, this Exploratory data analysis(EDA) helps in understanding IPL match trends and shows how data analysis and visualization can be used to extract meaningful information from sports data.

```python
team_runs = (
    df.groupby('batting_team')['runs_off_bat']
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

plt.figure(figsize=(6,4))
plt.bar(team_runs.index, team_runs.values)
plt.xlabel("Teams")
plt.ylabel("Runs")
plt.title("Top 5 Batting Teams")
plt.xticks(rotation=45)
plt.show()

```

This bar graph compares the total runs scored by the top 5 batting teams in the dataset.
X-axis: Teams
Y-axis: Total runs
Each bar represents the overall batting contribution of a team. Teams with higher total runs generally show better batting depth and consistency.
This visualization helps identify top-performing teams and supports performance comparison. It reveals that India is the most dominant batting team, followed by Australia, while Pakistan has the lowest total runs among the top five teams.

```python
# Runs per match
runs_per_match = df.groupby('match_id')['runs_off_bat'].sum()
# Balls per match
balls_per_match = df.groupby('match_id').size()
plt.figure(figsize=(8,5))
plt.scatter(balls_per_match, runs_per_match)
plt.xlabel("Total Balls")
plt.ylabel("Total Runs")
plt.title("Runs vs Balls per Match")
plt.show()
```

The scatter plot shows the relationship between total balls and total runs per match.This shows a positive correlation between balls and runs.As the number of balls increases, the total runs also increase. here ,
 Each dot = one match,
X-axis (horizontal) ‚Üí Total balls bowled in the match,
Y-axis (vertical) ‚Üí Total runs scored in the match. Teams that face more balls generally score more runs.
A few scattered points show matches where teams scored fewer runs despite many balls.

```python
noball_per_match = df.groupby('match_id')['noballs'].sum()
plt.figure(figsize=(10,4))
plt.plot(noball_per_match.index, noball_per_match.values)
plt.xlabel("Match ID")
plt.ylabel("No-balls")
plt.title("No-balls per Match")
plt.show()
```

We can see that ,No-balls range mostly between 0 to 2 per match
A few spikes up to 4 no-balls
Many matches have 0 or 1 no-ball
Pattern looks irregular, no steady trend

```python
# Select only numerical columns
num_df = df.select_dtypes(include=['int64', 'float64'])
num_df.head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(num_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of ICC Deliveries Data")
plt.show()
```

from the above insight we got to know that , Runs off the bat have a strong correlation with total runs.
Extras such as wides and no-balls also contribute to overall scoring.
Wickets show weaker correlation with runs, indicating wickets alone don‚Äôt decide total score.
The heatmap makes relationships between variables easy to understand.
Strong colors highlight important contributing factors to match outcomes.

```python
wickets = df[df['player_dismissed'] != 'Not Out']
top_bowlers = wickets['bowler'].value_counts().head(10)
plt.figure()
plt.bar(top_bowlers.index, top_bowlers.values,color='pink')
plt.xlabel('Bowler')
plt.ylabel('Wickets')
plt.title('Top 10 Wicket Takers in ICC')
plt.xticks(rotation=45)
plt.show()
```

We can observe that Competition among top bowlers is very tight
Performance is consistent across players
Success is based on long-term consistency, not one-time performance
Both fast bowlers and spinners appear ‚Üí balanced bowling attack . so A Zampa is top 1 bowler who has taken most of the wickets in the innings and followed by kuldeep yadav,jj bumrah and etc.

```python
over_consistency = df.groupby((df['ball'] - 1) // 6 + 1)['runs_off_bat'].std()

plt.figure(figsize=(8,5))
plt.plot(over_consistency.index, over_consistency.values,color='purple')
plt.xlabel("Over")
plt.ylabel("Run Variability (Std Deviation)")
plt.title("Over-wise Run Consistency (Without Creating Over Column)")
plt.show()
```

the line graph showing Runs increase in later overs, especially towards the death overs.
Early overs show steady scoring, while middle overs are more controlled.
Final overs have higher run rates due to aggressive batting.
This reflects common match strategy in limited-overs cricket.

```python
top_consistent = (
    df.groupby('striker')['runs_off_bat']
      .std()
      .dropna()
      .sort_values()
      .head(10)
)

plt.figure(figsize=(10,5))
plt.bar(top_consistent.index, top_consistent.values,color='violet')
plt.xlabel("Batsman")
plt.ylabel("Run Variability (Std Dev)")
plt.title("Top 10 Most Consistent Batsmen")
plt.xticks(rotation=45)
plt.show()
```

so we came to know that batsmen with lower variability are more consistent performers.
Consistent batsmen contribute reliable runs in most matches.
High variability indicates unpredictable performance.
This helps identify dependable players. Mahedi Hasan is the top 1 connsistent batsmen in the whole top 10 followed by SM Curran , Shariz Ahmad , R klein etc


```python
fours = (df['runs_off_bat'] == 4).sum()
sixes = (df['runs_off_bat'] == 6).sum()

plt.pie([fours, sixes], labels=['Fours', 'Sixes'], autopct='%1.1f%%')
plt.title("Boundary Distribution")
plt.show()
```

the pie chart is used to see th share of boundary by different teams to view how many teams striked sixes and fours. Players prefer ground shots over risky big hits , Sixes are less frequent, showing controlled batting
Boundaries mostly come from timing, not power

```python
plt.boxplot(df['runs_off_bat'])
plt.title("Runs per Ball Distribution")
plt.ylabel("Runs")
plt.show()
```

Batsmen usually play safe shots (dot balls & singles)
Boundaries are rare compared to normal runs
Scoring is not evenly spread ‚Üí a few big shots boost total runs
The distribution is right-skewed (many low values, few high values)

```python
wickets = df['wicket_type'].notna().astype(int)

wicket_over = wickets.groupby((df['ball'] - 1) // 6 + 1).sum()

plt.plot(wicket_over.index, wicket_over.values,color='green')
plt.title("Wickets per Over")
plt.xlabel("Over")
plt.ylabel("Wickets")
plt.show()
```

we can come to conclusion that Middle overs are more dangerous for batsmen
Bowlers are more effective when batsmen try to attack
Early overs are safer
Wicket-taking is not uniform across overs

```python
boundary_stats = df[df['runs_off_bat'].isin([4,6])].groupby(
    ['batting_team','runs_off_bat']
).size().unstack(fill_value=0)

top_teams = boundary_stats.sum(axis=1).sort_values(ascending=False).head(5).index
boundary_stats = boundary_stats.loc[top_teams]

boundary_stats.plot(kind='bar', figsize=(9,5))
plt.xlabel("Team")
plt.ylabel("Boundary Count")
plt.title("Top Teams ‚Äì Fours vs Sixes")
plt.xticks(rotation=45)
plt.show()
```

the visualization of the above bar graph is , Some teams score more boundaries, showing aggressive batting styles.
Teams with more sixes tend to have higher strike rates.
Fours indicate consistent boundary scoring, while sixes show power hitting.
This graph highlights attacking teams.
it also tells that higher four's have been striked by team australia and higher sixes have been striked by team south africa  

```python
team_fours = (
    df[df['runs_off_bat'] == 4]
    .groupby('batting_team')
    .size()
    .sort_values(ascending=False)
    .head(5)
)

plt.figure(figsize=(8,5))
plt.bar(team_fours.index, team_fours.values,color='green'),
plt.xlabel("Team")
plt.ylabel("Number of Fours")
plt.title("Top 5 Teams with Most Fours")
plt.xticks(rotation=45)
plt.show()
```

```python
team_sixes = (
    df[df['runs_off_bat'] == 6]
    .groupby('batting_team')
    .size()
    .sort_values(ascending=False)
    .head(5)
)

plt.figure(figsize=(8,5))
plt.bar(team_sixes.index, team_sixes.values,color='red')
plt.xlabel("Team")
plt.ylabel("Number of Sixes")
plt.title("Top 5 Teams with Most Sixes")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.barh(top_bowlers.index, top_bowlers.values, color='orange')
plt.xlabel("Number of Matches")
plt.ylabel("Bowler")
plt.title("Top 10 Bowlers by Matches Played")
plt.show()
```

```python
team_sr = df.groupby('batting_team')['runs_off_bat'] \
            .apply(lambda x: (x.sum()/x.count())*100) \
            .sort_values(ascending=False)

plt.figure(figsize=(10,5))
plt.plot(team_sr.values, marker='o')
plt.xticks(range(len(team_sr)), team_sr.index, rotation=45)
plt.xlabel("Team")
plt.ylabel("Strike Rate")
plt.title("Team Strike Rate ‚Äì Line Plot")
plt.grid(True)
plt.show()
```

this helps us to understand that , New Zealand has the highest strike rate, indicating very aggressive batting.
India and South Africa maintain high strike rates, showing a balance between aggression and consistency.
Teams like Bangladesh and Netherlands have lower strike rates, suggesting slower scoring.
The downward trend shows variation in batting tempo among teams.
This graph helps compare how quickly teams score runs.
